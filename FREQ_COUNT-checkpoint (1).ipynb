```python
import re
from collections import Counter
def count_word_occurrences(sentences, words):
    word_counts = {word: 0 for word in words}
    for sentence in sentences:
        tokens = re.findall(r'\b\w+\b', sentence.lower())
        for word in words:
            if word.lower() in tokens:
                word_counts[word] += 1
    return word_counts
# Example sentences
sentences = l2
# List of words to count occurrences
word_list = a
# Count occurrences of words
word_counts2 = count_word_occurrences(sentences, word_list)

# Print the table
print("| Word       | Count |")
print("|------------|-------|")
for word, count in word_counts2.items():
    print(f"| {word:<10} | {count:^5} |")
```

```python
# Tokenize, remove stopwords, punctuation, and perform stemming
def preprocess_sentence(sentence):
    stop_words = set(stopwords.words('english'))
    punctuation = set(string.punctuation)
    stemmer = PorterStemmer()
    tokens = re.findall(r'\b\w+\b', sentence.lower())
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.lower() not in punctuation]
    stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
    return stemmed_tokens

# Get unique words from all sentences
all_words = set()
for sentence in df['Sentence']:
    all_words.update(preprocess_sentence(sentence))

# Count word occurrences for positive, negative, and neutral sentiments
word_counts2 = count_word_occurrences(df[df['Sentiment']=='positive']['Sentence'], all_words)
word_counts3 = count_word_occurrences(df[df['Sentiment']=='negative']['Sentence'], all_words)
word_counts4 = count_word_occurrences(df[df['Sentiment']=='neutral']['Sentence'], all_words)

# Create DataFrames for word counts
table2 = pd.DataFrame(word_counts2.items(), columns=['Word2', 'Count2'])
table3 = pd.DataFrame(word_counts3.items(), columns=['Word3', 'Count3'])
table4 = pd.DataFrame(word_counts4.items(), columns=['Word4', 'Count4'])

# Combine word counts into a single DataFrame
final = pd.DataFrame()
final['word'] = table2['Word2']
final['pos_freq'] = table2['Count2']
final['neg_freq'] = table3['Count3']
final['neut_freq'] = table4['Count4']

# Predict sentiment for each sentence
def predict_sentiment(sentence):
    tokens = preprocess_sentence(sentence)
    positive_freq_total = sum(word_counts2.get(word, 0) for word in tokens)
    negative_freq_total = sum(word_counts3.get(word, 0) for word in tokens)
    neutral_freq_total = sum(word_counts4.get(word, 0) for word in tokens)
    if positive_freq_total > negative_freq_total and positive_freq_total > neutral_freq_total:
        return 'positive'
    elif negative_freq_total > positive_freq_total and negative_freq_total > neutral_freq_total:
        return 'negative'
    else:
        return 'neutral'

# Predict sentiment for all sentences
predictions = []
for sentence in df['Sentence']:
    predictions.append(predict_sentiment(sentence))

# Create DataFrame for actual and predicted values
results_df = pd.DataFrame({'Actual': df['Sentiment'], 'Predicted': predictions, 'Sentence': df['Sentence']})

# Display the DataFrame
print(results_df)
```
